```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sample_stats.csv")
```

```python
df
```

```python
n = df['idx'].unique()
n
```

```python
x_mewLi = []
for i in n:
    x_mewLi.append(df[df.idx == i]['x'].mean())
    print("x_mew", i, f" :{df[df.idx == i]['x'].mean()}")
print(x_mewLi)
```

```python
y_mewLi = []
for i in n:
    y_mewLi.append(df[df.idx == i]['y'].mean())
    print("y_mew", i, f" :{df[df.idx == i]['y'].mean()}")
print(y_mewLi)
```

```python
x_stdLi = []
for i in n:
    x_stdLi.append(df[df.idx == i]['x'].std())
    print("x_std", i, f" :{df[df.idx == i]['x'].std()}")
print(x_stdLi)
```

```python
y_stdLi = []
for i in n:
    y_stdLi.append(df[df.idx == i]['y'].std())
    print("y_std", i, f" :{df[df.idx == i]['y'].std()}")
print(y_stdLi)
```

```python
n.tolist()
```

```python
n = df['idx'].unique()
fig, axs = plt.subplots(3, 5, figsize = (16, 16))
for i, n in enumerate(n.tolist()):
    ax = axs[i]
    x = df[df.idx == n]['x']
    y = df[df.idx == n]['y']
    ax.plot(x, y)
```
